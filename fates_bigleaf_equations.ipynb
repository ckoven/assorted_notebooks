To solve for population number density and tree size of a mean-individual cohort given changes to the index biomass pool, we have two equations and two unknowns:
 
$n C(D) = A$

$n B(D) = M$

Where n is the number density, $C(D)$ is the crown area allometry as a function of stem diameter $D$, $A$ is the total patch area, $B(D)$ is the biomass allometry as a function of $D$, and $M$ is the cohort-aggregated biomass of the index biomass pool (probably stemwood). There is an analytical solution for the special case where both allometries are a power law:

$n C_1 D^{C_2} = A$

$n B_1 D^{B_2} = M$

Where n is the number density, $C_1$ and $C_2$ define a power law relationship of tree crown area as a function of stem diameter $D$, and $B_1$ and $B_2$ define a power law relationship of stem biomass as a function of $D$.

First solve for $D$ as a function of $n$ in the crown area allometry equation:

$D = \left(\frac{A}{n C_1}\right)^{\left(\frac{1}{C_2}\right)}$

Then substitute that back into biomass allometry equation to solve for $n$:

$n B_1 \left(\frac{A}{n C_1}\right)^{\left(\frac{B_2}{C_2}\right)} = M $

$n \left(\frac{1}{n}\right)^{\left(\frac{B_2}{C_2}\right)} \left(\frac{A}{C_1}\right)^{\left(\frac{B_2}{C_2}\right)} = \frac{M}{B_1} $

$n^{\left(1-\frac{B_2}{C_2}\right)} = \frac{M}{B_1} \left(\frac{A}{C_1}\right)^{\left(-\frac{B_2}{C_2}\right)} $

$n = \left( \frac{M}{B_1} \left(\frac{A}{C_1}\right)^{\left(-\frac{B_2}{C_2}\right)} \right) ^ { \frac{1}{\left(1 - \frac{B_2}{C_2}\right) }} $


Knowing $n$, then solve back for $D$ using either of the allometry equations.
